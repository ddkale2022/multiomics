```python
import pandas as pd
import numpy as np

# 1) Load & log‐transform (as before)
df = pd.read_csv('data_input/mrna.csv', index_col='id').T
logfpkm = np.log2(df + 0.1)

# 2) Gene‐level expression filter: require FPKM ≥1 in ≥50% of samples
#    (which on log2 scale is ≥ log2(1.1) ≈ 0.137, but we can just threshold on raw df)
keep_expr = (df >= 1.0).sum(axis=0) >= (0.5 * df.shape[0])
filtered_expr = logfpkm.loc[:, keep_expr]

print(f"Genes passing FPKM ≥1 in ≥50% samples: {filtered_expr.shape[1]} / {df.shape[1]}")

# 3) Low‐variance filter: drop bottom 10% of genes by variance
gene_vars = filtered_expr.var(axis=0)
cutoff = gene_vars.quantile(0.10)
keep_var = gene_vars > cutoff
filtered_final = filtered_expr.loc[:, keep_var]
filtered_final.to_csv('data_output/mrna_qc.csv', index=True)
mrna_qc    = pd.read_csv('data_output/mrna_qc.csv',    index_col=0)

print(f"Genes after variance filter (bottom 10% removed): {filtered_final.shape[1]}")

# 4) Your final mRNA matrix:
#    - samples = filtered_final.index (should still be 100)
#    - genes   = filtered_final.columns (e.g. ~8k–12k depending on stringency)

# Optional: inspect remaining gene count
remaining_genes = filtered_final.shape[1]
```

```python
import pandas as pd

# 1) Load & orient
df_proteo = pd.read_csv('data_input/proteo.csv', index_col='id').T
df_proteo = df_proteo.apply(pd.to_numeric, errors='coerce')

# 2) Sample‐level missingness filter
#    Keep samples with ≤30% missing intensities
keep_samples = (df_proteo.isna().mean(axis=1) <= 0.30)
proteo_step1 = df_proteo.loc[keep_samples]

# 3) Feature‐level missingness filter
#    Keep proteins quantified in ≥70% of retained samples
keep_proteins = (proteo_step1.isna().mean(axis=0) <= 0.30)
proteo_step2 = proteo_step1.loc[:, keep_proteins]

# 4) Low‐variance filter
#    Remove bottom 10% of proteins by variance
vars_prot = proteo_step2.var(axis=0, skipna=True).fillna(0)
cut_prot = vars_prot.quantile(0.10)
proteo_qc = proteo_step2.loc[:, vars_prot > cut_prot]

print("Proteome QC shape:", proteo_qc.shape)
```

```python
# 1) Load & orient
df_phospho = pd.read_csv('data_input/phospho.csv', index_col='id').T
df_phospho = df_phospho.apply(pd.to_numeric, errors='coerce')

# 2) Sample‐level missingness filter
#    Keep samples with ≤50% missing intensities
keep_samples = (df_phospho.isna().mean(axis=1) <= 0.50)
phospho_step1 = df_phospho.loc[keep_samples]

# 3) Feature‐level missingness filter
#    Keep sites quantified in ≥50% of retained samples
keep_sites = (phospho_step1.isna().mean(axis=0) <= 0.50)
phospho_step2 = phospho_step1.loc[:, keep_sites]

# 4) Low‐variance filter
#    Remove bottom 10% of phosphosites by variance
vars_phos = phospho_step2.var(axis=0, skipna=True).fillna(0)
cut_phos = vars_phos.quantile(0.10)
phospho_qc = phospho_step2.loc[:, vars_phos > cut_phos]

print("Phosphoproteome QC shape:", phospho_qc.shape)
```

```python
from mofapy2.run.entry_point import entry_point

# 0) (Optional) Prefix columns to guarantee uniqueness across views:
rna = mrna_norm_sub.copy()
rna.columns  = ["RNA__"  + c for c in rna.columns]

prot = prot_imp_sub.copy()
prot.columns = ["Prot__" + c for c in prot.columns]

phos = phos_imp2.copy()
phos.columns = ["Phos__" + c for c in phos.columns]

# 1) Build data_views as list-of-views, each view a list of group matrices:
#    Here we have a single group “all,” so each list has one element:
data_views = [
    [rna .values],   # shape = 72 × n_genes
    [prot.values],   # shape = 72 × n_proteins
    [phos.values]    # shape = 72 × n_phosphosites
]

# 2) One likelihood per view
likelihoods = ["gaussian", "gaussian", "gaussian"]

# 3) View names
views_names = ["RNA","Proteome","Phospho"]

# 4) Single group
groups_names = ["all"]

# 5) Sample names: a list containing one list of your 72 samples
samples_names = [rna.index.tolist()]

# 6) Feature names: one list per view
features_names = [
    rna .columns.tolist(),
    prot.columns.tolist(),
    phos.columns.tolist()
]

# 7) Load into MOFA2
ent = entry_point()
ent.set_data_matrix(
    data           = data_views,
    likelihoods    = likelihoods,
    views_names    = views_names,
    groups_names   = groups_names,
    samples_names  = samples_names,
    features_names = features_names
)

# 8) Configure & run
ent.set_data_options(scale_views=False)
ent.set_model_options(factors=10)
ent.set_train_options(maxiter=1000, convergence_mode='slow')
ent.build()
ent.run()
```

```python
# install if you don’t have it:
# pip install mygene

from mygene import MyGeneInfo
mg = MyGeneInfo()

def map_with_mygene(ids):
    out = mg.querymany(
        ids,
        scopes      = 'uniprot',      # treat input as UniProt IDs
        fields      = 'symbol',       # ask for HGNC symbol
        species     = 'human',
        as_dataframe=True
    )
    # filter successful mappings
    df = out.loc[out['symbol'].notna()]
    return df['symbol'].tolist()

# strip prefixes and map
top100_prot_un  = [p.split("__",1)[1] for p in top100_prot]
top100_phos_un  = [p.split("__",1)[1] for p in top100_phos]

prot_symbols = map_with_mygene(top100_prot_un)
phos_symbols = map_with_mygene(top100_phos_un)

print(f"Proteome mapped symbols: {len(prot_symbols)}/{len(top100_prot_un)}")
print(f"Phospho mapped symbols:  {len(phos_symbols)}/{len(top100_phos_un)}")
print("Example Proteome symbols:", prot_symbols[:10])
print("Example Phospho symbols: ", phos_symbols[:10])
```



## 1. **Data acquisition**

* **Sources**: Supplementary Tables S3–S5 from the LUAD multi-omics paper
* **Files**:

  * `mrna.csv` (RNA-Seq FPKM for 100 samples)
  * `proteo.csv` (iBAQ protein abundances for 206 runs)
  * `phospho.csv` (MaxQuant phosphosite intensities for 158 runs)
  * `metadata.csv` (sample labels: Tumor vs NAT)

## 2. **Quality control (QC)**

*For each modality (mRNA, proteome, phospho), working on samples × features matrices:*

1. **Sample filtering**: remove any sample with > 30 % missing values
2. **Feature filtering**: remove any gene/protein/site with > 30 % missing values
3. **Low-variance filter**: drop the bottom 10 % of features by variance
4. **Resulting QC shapes**:

   * mRNA: 100 × 16 188 → 100 × 13 319
   * Proteome: 206 × 11 119 → 182 × 6 154
   * Phospho: 158 × 22 564 → 158 × 5 014

## 3. **Complete-case intersection**

* Identified the **72 samples** profiled in all three QC’d tables
* Subset each QC matrix to those 72 samples
* Final QC shapes:

  * mRNA: 72 × 13 319
  * Proteome: 72 × 6 154
  * Phospho: 72 × 5 014

## 4. **Normalization & imputation**

* **mRNA**: log₂(FPKM + 0.1), quantile-normalize across samples
* **Proteome & Phospho**:

  1. log₂(intensity + 1)
  2. median-center each sample
  3. replace zeros with NaN and k-NN impute (k=5)
* Final processed shapes (72 samples × same feature counts)

## 5. **Exploratory PCA & clustering**

* Concatenated the three processed matrices (or ran PCA on each separately)
* Plotted the first two PCs and hierarchical clustering
* **Color‐coded** by Tumor vs NAT to confirm separation

## 6. **Extract & interpret top loadings (PC1)**

* Took the **100 features** with highest absolute loading on PC1
* Ran ORA (Enrichr) on those top features → strong “extracellular matrix organization” signal
* Heatmapped those features across the 72 samples → clear ECM-high vs ECM-low tumor subgroup

## 7. **Multi-omic factor analysis with MOFA2**

* Loaded the three processed matrices into MOFA2’s `entry_point`
* Set 10 latent factors, trained until ELBO convergence
* Extracted:

  * **Z** (72 samples × 10 factors)
  * **W** for each view (features × 10 factors)
* Confirmed Factor 1 perfectly separates Tumor vs NAT (p ≈ 10⁻³⁹, AUC = 1.0)

## 8. **Factor interpretation**

* **Factor 1** captures the joint proliferation + ECM/remodeling program
* **Factor 2** top drivers (RNA, Prot, Phos) enrich for endocytosis, focal adhesion, phosphoinositide metabolism → a membrane/cytoskeleton remodeling axis
* All other factors (3–10) show no Tumor/NAT signal, but can be mined for immune, metabolic or other biology

## 9. **Downstream modeling & visualization**

* **Tumor vs NAT classifier**: logistic regression on Z → AUC = 1.0 with Factors 1 ± 2
* **Unsupervised subtyping**: K-means (k=2–3) in Factor 1–3 space to define molecular subgroups
* **Heatmaps**: combined top drivers from Factor 1 & 2 across all three omics
